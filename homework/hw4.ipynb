```python
from matplotlib import pyplot as plt
import numpy as np
from numpy import linspace, zeros, absolute

iterations = 50
resolution = 1000

x_set = linspace(-2,0.5,300)
y_set = linspace(-1,1,300)

a1,b1,p1 = -2,0.5,300  
a2,b2,p2 = -1,1,300 


#make an array
mandset = zeros([p1,p2],float)


#calculate the array
for i in range(len(y_set)): 
    for j in range(len(x_set)):
        
        #initial conditions
        z = 0        
        c = x_set[j] + y_set[i]*1j       
        inset = True

        #we need to iterate it to see if the point satisfies the condition
        for k in range(iterations):
            z = z**2+c
            #print('values:  ', k,z,c)
            if absolute(z)>2:
                mandset[i,j]=k
                inset = False
                break




width = 300
height = 300

mandelbrot = np.empty((width,height))


#plot
plt.figure(figsize=(100,100))
plt.imshow(mandelbrot,origin="lower",extent=[-2,2,-2,2],cmap="hot")

plt.show()
```
