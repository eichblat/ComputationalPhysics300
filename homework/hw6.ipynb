```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt


#define function
f = lambda x, a, b: a * np.exp(-(x**2))

#parameters
xmin, xmax = -5, 5
points = 100
order1 = 2
order2 = 4
a = 1
b = 2

#arrays
x = np.linspace(xmin, xmax, points)
y = f(x, a, b) + np.random.rand(points)- 0.5

fita = np.polyfit(x, y, order1, full=True)
yp1 = np.polyval(fita[0], x)

fitb = np.polyfit(x, y, order2, full=True)
yp2 = np.polyval(fitb[0], x)



#plot the data
plt.plot(x, y, 'r', label = 'noisy data')
plt.plot(x, f(x, a, b), label = 'ideal function') 
plt.plot(x, yp1, label = 'fita')
plt.plot(x, yp2, label = 'fitb') 



plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.xlim([xmin, xmax])
plt.legend(fontsize=10)

```
